# 1. Import Data

```python
import pandas as pd

file_url = 'https://raw.githubusercontent.com/ajayt0m/Telecom_Dashboard_ConnectX/main/A-1_NO_OF_VILLAGES_TOWNS_HOUSEHOLDS_POPULATION_AND_AREA.xlsx'

census_df = pd.read_excel(file_url)
```

```python
census_df
```

# 2. Cleaning Header

```python
census_df.head()
```

```python
for col in census_df.columns:
    print(col)
```

```python
#Rename 'India/ State/ Union Territory/ District/ Sub-district' to 'Geographic Subdivision'
census_df.rename(columns = {'India/ State/ Union Territory/ District/ Sub-district':'Geographic Subdivision'}, inplace = True)

#Rename 'Total/\nRural/\nUrban' to 'Total/Rural/Urban'
census_df.rename(columns = {'Total/\nRural/\nUrban':'Total/Rural/Urban'}, inplace = True)

#Rename 'Number of villages' to 'NV_Inhabited'
census_df.rename(columns = {'Number of villages':'NV_Inhabited'}, inplace = True)

#Rename 'Unnamed: 7' to 'NV_Uninhabited'
census_df.rename(columns = {'Unnamed: 7':'NV_Uninhabited'}, inplace = True)

#Rename 'Population' to 'P_Total'
census_df.rename(columns = {'Population':'P_Total'}, inplace = True)

#Rename 'Unnamed: 11' to 'P_Males'
census_df.rename(columns = {'Unnamed: 11':'P_Males'}, inplace = True)

#Rename 'Unnamed: 12' to 'P_Females'
census_df.rename(columns = {'Unnamed: 12':'P_Females'}, inplace = True)

#Rename 'Area\n (In sq. km)' to 'Area(sq.km)'
census_df.rename(columns = {'Area\n (In sq. km)':'Area(sq.km)'}, inplace = True)
```

```python
census_df.head()
```

```python
for col in census_df.columns:
    print(col)
```

# 3. Delete first 2 rows - Useless values

```python
census_df_hc = census_df.iloc[2:, :]
```

```python
census_df_hc.head()
```

# 4. Removing Junk values

```python
import numpy as np
```

```python
#making junk values in 'Distric Code' to nan
junk_values = ['@', '&', '$', '###', '#', '^', '*', '**', '##', '^^', 'Note:', '1', '2', '3', '4', '(ii)', '5', 'NA -']
census_df_hc['District Code'] = census_df_hc['District Code'].replace(junk_values, np.nan)
```

```python
#making junk values in 'Geographic Subdivision' to nan
junk_values = ['total',7687.39, 10137.61]
census_df_hc['Geographic Subdivision'] = census_df_hc['Geographic Subdivision'].replace(junk_values, np.nan)
```

```python
#making junk values in 'Total/Rural/Urban' to nan
junk_values = [106.16, 309.35]
census_df_hc['Total/Rural/Urban'] = census_df_hc['Total/Rural/Urban'].replace(junk_values, np.nan)
```

```python
import matplotlib.pyplot as plt
```

# 5. Missing Values

```python
plot_missing = census_df_hc.isna().sum().plot(kind="bar")
plot_missing.set_xlabel("Columns", fontsize=12)
plot_missing.set_ylabel("Count of missing values", fontsize=12)
plot_missing.set_title("Missing Values per Column", fontsize=14)
plot_missing.set_xticklabels(census_df_hc.columns, rotation=45, ha="right")
plt.show()
```

```python
#Rows with nan for more than 8 columns would be dropped
#28 rows dropped
census_df_hc = census_df_hc.dropna(thresh=census_df_hc.shape[1]-8)
census_df_hc
```

# 6. Imputation

```python
#Need to impute 117 rows where the 2 column values are nan

census_clean = census_df_hc
census_clean['Area(sq.km)']=census_clean['Area(sq.km)'].fillna(census_clean['Area(sq.km)'].mean())
census_clean.info()
```

```python
census_clean['Population per sq. km.']=census_clean['Population per sq. km.'].fillna(census_clean['Population per sq. km.'].mean())
census_clean.info()
```

### Checking for any row with atleast one null value

```python
census_clean[census_clean.isnull().any(axis=1)]
```

```python
plot_missing = census_clean.isna().sum().plot(kind="bar")
plot_missing.set_xlabel("Columns", fontsize=12)
plot_missing.set_ylabel("Count of missing values", fontsize=12)
plot_missing.set_title("Missing Values per Column", fontsize=14)
plot_missing.set_xticklabels(census_clean.columns, rotation=45, ha="right")
plt.show()
```

# 7. Further cleaning - Junk Values

### Need to change 'JAMMU & KASHMIR @&' to 'JAMMU & KASHMIR'
### Need to change 'INDIA @&', 'INDIA $' to 'INDIA

```python
census_clean['Name'] = census_clean['Name'].str.replace('@&', '')
census_clean['Name'] = census_clean['Name'].str.replace('$', '')
```
